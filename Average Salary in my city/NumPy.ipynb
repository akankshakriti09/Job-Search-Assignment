Average Salary for Python Developers in my City (Patna, Bihar)

```python
import pandas as pd
```

```python
df = pd.read_csv("WebScraping/cleaned_jobList.csv",encoding="utf-8")
```

```python
df.head()
```

```python
specific_columns=df[["company_location","avg_salary"]]
print(specific_columns)
```

```python
specific_columns.head(30)
```

```python
find_avg_salary = df[df["company_location"] == "Patna, Bihar"]
find_avg_salary
```

Average salary of specific location using mean function

```python
sal = find_avg_salary["avg_salary"].mean()
print("Average Salary of my city (Patna, Bihar) is Rs.",sal)
```

Average salary of specific location using Numpy

```python
import numpy as np
array = np.array(find_avg_salary["avg_salary"])
a = np.average(array)
print("The average salary of my city (Patna, Bihar) is Rs.", abs(round(a, 3)))
 
# Output:
# The salary of my city (Patna, Bihar) is Rs. 18900.0
```
