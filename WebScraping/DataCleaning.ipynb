```python
import pandas as pd
```

```python
df = pd.read_csv("jobList_results.csv",encoding="utf-8")
```

```python
def format_salary(row):
    Salary = row["salary"]
    row["salary"]= Salary.replace("â‚¹","").replace("a month","").replace("a year","").replace(",","").replace("From","").replace("Full-time","10000").replace("Part-time +3","10000").replace("Up to","")
    
    if "-" in row["salary"]:
        split = row["salary"].split("-")
        salary_min = split[0]
        salary_max = split[1]
    else:
        salary_min = row["salary"]
        salary_max = row["salary"]
        
    
    # row["old_salary"] = Salary
    row["salary_min"] = salary_min 
    row["salary_max"] = salary_max 
    
       
    if "month" in row["salary"]:
        row["salary_min"] = int(row["salary_min"])*12
        row["salary_max"] = int(row["salary_max"])*12
   
    return row

df_salary = df[df["salary"]!= "None"].fillna("12000") #replacing null values in salary with 12000
df_salary = df_salary.apply(format_salary,axis=1) 
df_salary['avg_salary'] = (df_salary["salary_min"].astype(int) + df_salary["salary_max"].astype(int))/2
```
